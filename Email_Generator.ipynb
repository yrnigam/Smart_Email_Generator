### Task 1 : Fetch the Detail like name, email, email Body, email subject and create email template for each individuals. 

```python
#Importing data from csv file
import csv
mail=[]
with open("./sample_email.csv", 'r') as file:
    csv_file = csv.DictReader(file)
    for row in csv_file:
        mail.append(dict(row))
```

```python
#Fetch the Detail for each individual
sender= "Tony Stark"
newline = '\n'
all_mails=[]
for i in range(len(mail)):
    first_name = mail[i].get('first_name')
    last_name = mail[i].get('last_name')
    email = mail[i].get('email')
    subject = mail[i].get('subject')
    body = mail[i].get('Email Boby')
    
    #create email template for each individuals
    final_mail = (
    f"Email to : {email} {newline}" 
    f"Subject Line: {subject} {newline}"   
    f"Hi {first_name} {last_name}, {newline}" 
    f"{body} {newline}"   
    f"Thanks,{newline}{sender}")
    
    all_mails.append(final_mail)
    
```

```python
#Printing mail output for all individual 
for i in range(len(all_mails)):
    print(all_mails[i])
    print("==================================")
```

### Task 2 : Check and fetch email address from the email body section if any present and store it in another CSV. 

```python
#Importing regular expression
import re 
mail_list=[]

#Creating list of mails
for i in range(len(all_mails)):
    body = mail[i].get('Email Boby')
    email  =re.findall('\S+@\S+',body)
    if email != []:
        mail_list+=email

#Write list to CSV file
all_mail=",\n".join(mail_list)
f = open('email.csv','w')
f.write(all_mail)
f.close()
```

### Task 3 : Check and fetch mail number from the email body section if any present and store it in another CSV.

```python
contact_list=[]

#Creating list of contact number
for i in range(len(all_mails)):
    body = mail[i].get('Email Boby')
    phone  =re.findall('[\+\(]?[1-9][0-9 .\-\(\)]{8,}[0-9]',body)
    if phone != []:
        contact_list+=phone
        
#Write list to CSV file
all_contact=",\n".join(contact_list)
f = open('contact_number.csv','w')
f.write(all_contact)
f.close()
```
